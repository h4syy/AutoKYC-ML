```python
import os
from ultralytics import YOLO
import cv2
```

```python

# Load the trained YOLO model (best.pt or last.pt)
model = YOLO('C:/Users/sulav.adhikari/Desktop/output/best.pt')

# Path to the directory containing bulk images
image_dir = 'C:/Users/sulav.adhikari/Desktop/output/TESTING'
output_dir = './output/'

# Make sure output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop through all images in the directory
for image_name in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_name)
    
    # Perform inference on each image
    results = model.predict(source=image_path, save=True, save_txt=True, save_conf=True)
    
    # Save results to the output directory
    for result in results:
        output_image_path = os.path.join(output_dir, f"result_{image_name}")
        result.plot(save_path=output_image_path)  # Save output image with predictions
```
